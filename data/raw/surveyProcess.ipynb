```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/MindCare/surveyForm .csv")
df.head()
```

```python
df = df.dropna
```

```python
data = df[["Q23: In your own words, what mental health challenges are you currently facing? ","Q24: What would help you feel better or manage your mental health better? ","Q25: Is there anything else you'd like to share about your mental health or well-being? "]]
```

```python
data.isnull().sum()
```

```python

data = data.dropna()
```

```python
data.isnull().sum()
```

```python
melted = data.melt(value_name="user", var_name="question").dropna(subset=[("user")])
melted.head()
```

```python
melted["assistant"] = "I understand. It’s okay to feel this way. You’re not alone."
```

```python
import json

with open("processed/survey.jsonl", "w", encoding="utf-8") as f:
    for _, row in melted.iterrows():
        f.write(json.dumps({
            "user": row["user"],
            "assistant": row["assistant"]
        }, ensure_ascii=False) + "\n")

print("✅ survey.jsonl saved")
```
